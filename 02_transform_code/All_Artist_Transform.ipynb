## Merge 3 artists files into 1 table

```python
len(cmoa_artists)
```

```python
len(moma_artists)
```

```python
len(tate_artists)
```

```python
total_length = (len(cmoa_artists)+len(moma_artists)+len(tate_artists))
print(total_length)
```

```python
artists_df_list = [moma_artists, cmoa_artists, tate_artists]
artists_df = pd.concat(artists_df_list)
```

```python
artists_df.tail()
```

```python
len(artists_df)
```

```python
artists_df.dtypes
```

```python
# drop duplicate artist names
artists_df=artists_df.drop_duplicates(subset=['artist_name'])
len(artists_df)
```

```python
# drop artist_id columm, to then add back in and get unique values
artists_df = artists_df.drop(['artist_id'], axis =1)
```

```python
artists_df.insert(0,'artist_id', range(1,1+len(artists_df)))
artists_df.head()
```

```python
artists_df.tail()
```

```python
# sort by artist name
artists_df = artists_df.sort_values('artist_name')
artists_df
```

```python
# reset index
artists_df = artists_df.reset_index(drop=True)
artists_df
```

```python
# Remove first row with no artist name
artists_df = artists_df.iloc[1:]
artists_df
```

```python
# export to see results in csv
artists_df.to_csv('../03_load/artists.csv', header=True, index=False)
```
