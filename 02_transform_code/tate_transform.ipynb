```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Store filepath in variable
tate_data = "../00_input/the-tate-collection.csv"
```

```python
# Read data file into pandas
# https://stackoverflow.com/questions/24606330/how-to-read-a-file-with-a-semi-colon-separator-in-pandas
tate_df = pd.read_csv(tate_data, sep=';')
tate_df.head()
```

```python
list(tate_df.columns)
```

```python
# Filter columns for tate_artists_df
tate_artists_df = tate_df.filter([
    'artist',
    'artistRole'
])
list(tate_artists_df.columns)
```

```python
# Create a new column with just the first part before comma
tate_artists_df['last_name'] = tate_artists_df['artist'].str.split(', ').str[0]
tate_artists_df.head()
```

```python
# Create a new column with just the sescond part after comma
tate_artists_df['first_name'] = tate_artists_df['artist'].str.split(', ').str[1]
tate_artists_df.head()
```

```python
# CONCATENATE
tate_artists_df['artist_name'] = tate_artists_df['first_name'] + str(" ") + tate_artists_df['last_name']
tate_artists_df
```

```python
# Create copy of original df, keep only specific columns using '.filter' which will create a copy by default
# https://stackoverflow.com/questions/34682828/extracting-specific-selected-columns-to-new-dataframe-as-a-copy
tate_artists_df = tate_artists_df.filter(['artistRole','artist_name'])
tate_artists_df.head()
```

```python
list(tate_artists_df.columns)
```

```python
# Rename columns in tate_artists_df to align with list
tate_artists_df.columns = [
    'artist_role',
    'artist_name'
]
tate_artists_df.head()
```

```python
# Add column headers with empty values
tate_artists_df['artist_id']= ' '
tate_artists_df['nationality']= ' '
tate_artists_df['gender']= ' '
tate_artists_df['birth_year']= ' '
tate_artists_df['birth_place']= ' '
tate_artists_df['death_year']= ' '
tate_artists_df['death_place']= ' '
tate_artists_df['wiki']= ' '
tate_artists_df['ulan']= ' '
```

```python
list(tate_artists_df.columns)
```

```python
tate_artists_df.dtypes
```

```python
# Replace empty strings in artist_name with NaN
tate_artists_df['artist_name'].replace('', np.nan, inplace=True)
```

```python
# Drop rows with NaN in artist_name
tate_artists_df.dropna(subset=['artist_name'], inplace=True)
```

```python
# Drop duplicate in artist_name
tate_artists_df.drop_duplicates(subset ='artist_name',
                               keep=False, inplace=True)
tate_artists_df
```

```python
# Reorder tate columns
tate_artists_df = tate_artists_df[[
    'artist_name',
    'artist_role',
    'artist_id',
    'nationality',
    'gender',
    'birth_year',
    'birth_place',
    'death_year',
    'death_place',
    'wiki',
    'ulan']]
list(tate_artists_df.columns)
```

```python
# Sort by artist_name
tate_artists_df = tate_artists_df.sort_values('artist_name')
tate_artists_df
```

```python
# Reset index
tate_artists_df = tate_artists_df.reset_index(drop=True)
tate_artists_df
```

```python
# Replace 'Nan' with empty string
tate_artists_df = tate_artists_df.replace(np.nan, '', regex=True)
tate_artists_df
```

```python
tate_artists_df.to_csv(r'..\01_extract_tate\tate_artists.csv', index=False)
```
